# **PART A - Application Area Review**
**Literature review**

Artificial Intelligence has been used in medicine and healthcare for over fifty years, long before the recent growth of deep learning and large language models. Although the core ideas of AI emerged in the 1950s, early real-world applications focused on medical decision support systems designed to assist doctors rather than replace them (Shortliffe and Buchanan, 1975; Patel et al., 2009).

One of the earliest and most influential medical AI systems was MYCIN, developed at Stanford University in the early 1970s. MYCIN was a rule-based expert system created to diagnose bacterial infections and recommend appropriate antibiotics based on patient symptoms and laboratory results. Studies showed that MYCIN performed at a level comparable to trained medical professionals. However, it was never deployed in clinical practice due to concerns about explainability, system maintenance, and legal responsibility in the case of incorrect recommendations (Shortliffe, 1976; Buchanan and Shortliffe, 1984).

Around the same period, INTERNIST-I was developed as a diagnostic support system covering a wide range of diseases. It relied on manually encoded medical knowledge to suggest possible diagnoses based on observed symptoms. These early systems highlighted a pattern that still applies today: AI in medicine is primarily used to support clinical decision-making rather than make final decisions independently, due to the high risks associated with medical errors and patient safety concerns (Topol, 2019).

Early medical AI systems were largely based on rule-based expert systems, where knowledge was encoded using if–then rules created by domain experts. While effective for narrow and well-defined tasks, these systems had significant limitations. They were difficult to scale, required extensive manual effort to update, and performed poorly when encountering cases outside their predefined rules. Additionally, they struggled to adapt to new medical knowledge or changing clinical practices. As a result, many of these systems remained confined to research settings and were not widely adopted in hospitals. Limited computing power and the lack of large digital medical datasets further restricted their practical use during this time (Patel et al., 2009; Jiang et al., 2017).

The adoption of AI in medicine increased significantly in the 2000s with the rise of machine learning and deep learning techniques. Unlike rule-based approaches, machine learning models could learn patterns directly from data, reducing the need for manually written rules (Obermeyer and Emanuel, 2016). Deep learning models achieved strong performance in areas such as medical imaging analysis, disease prediction, and clinical risk assessment. These models were better suited to large datasets and showed improved generalisation, making AI systems more practical for real-world clinical deployment (Esteva et al., 2017; Rajkomar et al., 2019).

Despite these advances, new challenges emerged related to model transparency, reliability, and error behaviour, which are especially important in healthcare environments (Amann et al., 2020). Healthcare data also contains large amounts of unstructured text, including clinical notes, diagnostic reports, and biomedical literature. This has made Natural Language Processing (NLP) a key component of modern medical AI systems (Wang et al., 2018).

A notable milestone in medical NLP was IBM Watson’s DeepQA system, which aimed to answer clinical questions by analysing medical literature and patient data. This demonstrated the potential of AI to support evidence-based clinical decision-making using textual information (Ferrucci et al., 2013).

More recently, transformer-based models and large language models (LLMs) have been applied to tasks such as clinical question answering, medical summarisation, and decision support. These models can process large volumes of text and generate fluent responses, making them attractive for healthcare applications that rely heavily on language data (Lee et al., 2020; Singhal et al., 2023). However, their use has introduced a major risk known as hallucination, where models generate fluent but incorrect or unsupported information. In healthcare, such errors are particularly dangerous because they can negatively impact diagnosis, treatment decisions, and patient safety (Ji et al., 2023).

Unlike earlier systems, modern generative models may produce confident answers even when they are incorrect or uncertain. This has made hallucination detection and mitigation an important research focus in medical AI, especially for clinical decision support and medical question-answering systems (Maynez et al., 2020; Zhang et al., 2023). Recent studies show rapid growth in healthcare AI research alongside increasing concerns about reliability, bias, and ethical deployment (Tran et al., 2024).

Several approaches are being explored to reduce hallucinations in medical AI systems. These include retrieval-augmented generation to ground outputs in trusted medical sources, confidence scoring and uncertainty estimation to assess answer reliability, human-in-the-loop validation for high-risk outputs, and evaluation of generated responses against structured medical knowledge bases.

Overall, research is increasingly focused on developing trustworthy and explainable AI systems that are transparent, reliable, and aligned with clinical standards. As large language models continue to be integrated into healthcare settings, hallucination detection is expected to remain a critical area of research and development (Amann et al., 2020; Singhal et al., 2023).

This project focuses on trustworthy NLP-based medical decision support, with particular emphasis on hallucination detection in medical text generation. Understanding the evolution of medical AI from early rule-based systems to modern deep learning and language models provides essential context for why hallucination detection is necessary in contemporary healthcare applications.

# Technique 1: Retrieval-Augmented Generation (RAG)

***AI Paradigm: Search Techniques and Generative LLM***

**Description and Application**

Retrieval-Augmented Generation (RAG) is used to reduce hallucinations by grounding generated answers in real external knowledge sources. Before producing an answer, the system searches a large medical database, such as PubMed, using information retrieval or vector similarity methods. The retrieved documents are then passed to a language model, which generates an answer based on this evidence instead of relying only on its internal knowledge.
In medical question-answering, hallucinations often occur when a model generates answers without reliable facts. For example, when asked about a medical treatment, a language model may invent dosage details or side effects. RAG reduces this risk by forcing the model to use real medical documents. This makes it suitable for applications such as clinical decision support systems and medical chatbots where factual accuracy is critical.

**Input Data and Output**

Input: Medical question text and a large external medical document database.
Output: A generated medical answer grounded in retrieved documents, often with supporting evidence.

**Data Availability**

Large medical datasets such as PubMed are publicly available. However, they must be cleaned, split into smaller chunks, embedded, and indexed using tools like FAISS or ElasticSearch. This process requires significant time and expertise.

**Time to Set Up and Produce Results**

Setup time is high because both a search system and a generation model must be implemented and integrated. Once deployed, answers can be generated quickly, but overall system complexity remains high.

**Strengths and Weaknesses**

Strengths: Strong hallucination reduction, realistic for clinical systems, supports evidence-based medicine.
Weaknesses: Retrieval errors directly lead to incorrect answers, high engineering effort, difficult to evaluate retrieval quality.

**Failure Modes in Medical Settings**

If the system retrieves irrelevant or outdated documents, the generated answer may still be incorrect. RAG also struggles when no relevant documents exist for rare medical conditions.

**Why It Was Not Implemented**

This project focuses on detecting hallucinations after answers are generated. RAG operates during answer generation, so implementing it would change the project scope and exceed coursework complexity limits.


# Technique 2: Uncertainty Estimation using Stochastic Deep Learning

***AI Paradigm: Deep Learning and Probabilistic Reasoning (Bayesian-Inspired)***

**Description and Application**

Uncertainty estimation detects hallucinations by measuring how confident a language model is about its answers. Instead of checking facts directly, this approach looks at signals such as prediction entropy, variation across multiple generated answers, or disagreement between outputs.

In medical hallucination detection, this technique flags answers that appear unstable or inconsistent. For example, if the same medical question produces different answers across multiple runs, this suggests hallucination risk. Research models such as H-STO-TRANS introduce randomness into the transformer attention mechanism to expose hidden uncertainty that standard models do not reveal.

**Input Data and Output**

Input: Generated medical answers and internal probability outputs from a language model.
Output: An uncertainty or risk score indicating how unreliable an answer may be.

**Data Availability**

No external medical databases are required. However, access to a generative language model and the ability to run multiple stochastic passes are necessary.

**Time to Set Up and Produce Results**

Setup time is medium to high. Producing results is slower because multiple generations or forward passes are required for each question.

**Strengths and Weaknesses**

Strengths: Can detect fluent but incorrect answers, strong theoretical grounding, model-agnostic.
Weaknesses: Computationally expensive, difficult to interpret, hard to set reliable uncertainty thresholds.

**Failure Modes in Medical Settings**

Some hallucinated answers may still appear confident, meaning low uncertainty does not always guarantee correctness. This is dangerous in medical contexts where confident errors can cause harm.

**Why It Was Not Implemented**

This technique requires direct access to generative models and repeated sampling, which was beyond the computational and architectural scope of this coursework.

# Technique 3: Supervised Ensemble Learning (Gradient Boosting)

***AI Paradigm: Machine Learning (Supervised Learning / Ensemble Methods)***

**Description and Application**

This technique treats hallucination detection as a supervised classification problem. Instead of changing how answers are generated, it evaluates answers after they already exist. This is known as post-hoc detection.

Gradient Boosting combines several simple but effective signals:

Semantic similarity between the answer and medical knowledge (faithfulness)

Semantic similarity between the answer and the question (relevance)

Abnormal answer length compared to the reference knowledge

In medical hallucination detection, this approach identifies answers that do not align with known medical facts or do not properly address the question.

**Input Data and Output**

Input: Question text, generated answer, and reference medical knowledge converted into numerical features.
Output: A binary decision (hallucinated or factual) with a confidence score.

**Data Availability**

This approach requires labeled data. The MedHallu dataset provides medical questions, answers, supporting knowledge, and hallucination labels, making it suitable for supervised learning.

**Time to Set Up and Produce Results**

Setup time is low to medium using standard machine learning libraries. Training and inference are fast and run efficiently on CPUs.

**Strengths and Weaknesses**

Strengths: Easy to interpret, efficient, works well with structured features, suitable for safety-critical systems.
Weaknesses: Requires labeled data and does not explain the exact medical reasoning behind hallucinations.

**Failure Modes in Medical Settings**

If the dataset labels are biased or incomplete, the model may learn incorrect patterns. The system also depends on the quality of similarity features.

**Why This Technique Was Implemented**

Compared to RAG and uncertainty estimation, Gradient Boosting offers the best balance between performance, interpretability, and simplicity. It fits coursework constraints and directly supports post-hoc hallucination detection, making it ideal for Part C.

**When is this approach appropriate?**

This Gradient Boosting–based hallucination detection approach is appropriate when a medium-scale labelled dataset is available, explainability is required, and computational resources are limited. It is well suited for post-hoc validation of generated medical answers where the generation model cannot be modified.

This approach is not suitable for zero-label settings, multilingual generalisation, or scenarios that require end-to-end control of answer generation, as it only detects hallucinations after the response has been produced.

**Comparison**

RAG reduces hallucinations during answer generation but is complex and data-heavy.
Uncertainty estimation identifies risky answers using model confidence but is computationally expensive.
Gradient Boosting provides a simple, interpretable, and efficient post-hoc detection approach using labeled data.

Other research approaches, such as claim-level fact verification and large transformer-based self-checking models, exist but were not selected due to higher complexity and lack of suitable labeled data.

## Train, Validation, and Test Split

To prevent data leakage, the dataset is split **by unique medical questions**, not by individual rows.

This ensures that the same question never appears in multiple splits, even though each question has both a factual and a hallucinated answer.

The final split is:
- 72% training
- 8% validation
- 20% testing


```python
unique_questions = final_df["Question"].unique()

train_q, test_q = train_test_split(
    unique_questions, test_size=0.2, random_state=42
)
train_q, val_q = train_test_split(
    train_q, test_size=0.1, random_state=42
)

train_df = final_df[final_df["Question"].isin(train_q)]
val_df   = final_df[final_df["Question"].isin(val_q)]
test_df  = final_df[final_df["Question"].isin(test_q)]

print("Train:", train_df["label"].value_counts().to_dict())
print("Val:",   val_df["label"].value_counts().to_dict())
print("Test:",  test_df["label"].value_counts().to_dict())
```

## Feature Engineering using Sentence Embeddings

Text inputs are converted into numerical features using a pre-trained **SentenceTransformer (MiniLM)** model.

Instead of using raw embeddings directly, semantic similarity scores are computed to capture:
- Faithfulness to medical knowledge
- Relevance to the original question
- Abnormal verbosity patterns


```python
sb_model = SentenceTransformer("paraphrase-MiniLM-L3-v2")
```

```python
def length_ratio(a, b):
    return len(a.split()) / (len(b.split()) + 1e-6)
```

```python
train_q_emb = sb_model.encode(
    train_df["Question"].tolist(),
    batch_size=64,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
)

train_ans_emb = sb_model.encode(
    train_df["Answer"].tolist(),
    batch_size=64,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
)

train_kn_emb = sb_model.encode(
    train_df["Knowledge"].tolist(),
    batch_size=64,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
)
```

## Semantic Similarity Features

Three interpretable features are extracted for each answer:

1. **Answer–Knowledge similarity**  
   Measures how well the answer aligns with trusted medical evidence (faithfulness).

2. **Answer–Question similarity**  
   Measures whether the answer directly addresses the medical question (relevance).

3. **Length ratio**  
   Captures abnormal verbosity, which is a known signal of hallucination.


```python
# Answer–Knowledge (faithfulness)
train_sim = np.sum(train_ans_emb * train_kn_emb, axis=1)

# Answer–Question (relevance)
train_qa_sim = np.sum(train_ans_emb * train_q_emb, axis=1)

# Length abnormality
train_len_ratio = [
    length_ratio(a, k)
    for a, k in zip(train_df["Answer"], train_df["Knowledge"])
]
```

```python
X_train = np.column_stack([
    train_sim,        # Faithfulness
    train_qa_sim,     # Relevance (CR)
    train_len_ratio   # Length abnormality
])

y_train = train_df["label"].values
```

```python
test_q_emb = sb_model.encode(
    test_df["Question"].tolist(),
    batch_size=64,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
)

test_ans_emb = sb_model.encode(
    test_df["Answer"].tolist(),
    batch_size=64,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
)

test_kn_emb = sb_model.encode(
    test_df["Knowledge"].tolist(),
    batch_size=64,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
)

test_sim = np.sum(test_ans_emb * test_kn_emb, axis=1)
test_qa_sim = np.sum(test_ans_emb * test_q_emb, axis=1)

test_len_ratio = [
    length_ratio(a, k)
    for a, k in zip(test_df["Answer"], test_df["Knowledge"])
]

X_test = np.column_stack([
    test_sim,
    test_qa_sim,
    test_len_ratio
])

y_test = test_df["label"].values
```

## Gradient Boosting Classifier

A Gradient Boosting classifier is used to combine the semantic features into a final hallucination prediction.

Gradient Boosting was chosen because it:
- Handles non-linear feature interactions
- Works well with small, structured feature sets
- Provides strong performance without requiring GPUs


```python
clf = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred,
    display_labels=["Factual", "Hallucinated"],
    cmap="Blues"
)

plt.title("Confusion Matrix")
plt.show()
```

## Hyperparameter Tuning

Grid search with cross-validation is used to tune key Gradient Boosting parameters.

The F1-score is used as the evaluation metric to balance false positives and false negatives, which is important in medical safety settings.


```python
param_grid = {
    "n_estimators": [100, 200],
    "learning_rate": [0.05, 0.1],
    "max_depth": [2, 3]
}

grid = GridSearchCV(
    GradientBoostingClassifier(random_state=42),
    param_grid,
    scoring="f1",
    cv=3
)

grid.fit(X_train, y_train)
```

```python
# Get tuned model
best_clf = grid.best_estimator_

# Predict
y_pred = best_clf.predict(X_test)

# Evaluate
print("Best parameters:", grid.best_params_)
print(classification_report(y_test, y_pred))
```

**1. Verification of Correctness and Testing Methods**

First, the dataset was split into training, validation, and test sets using unique questions, rather than random rows. This ensures that answers related to the same question do not appear across different splits. This testing approach is important for medical question answering, where similar answers can easily cause data leakage. By splitting at the question level, the implementation correctly tests whether the model generalises to unseen medical questions, not just memorised patterns.

Second, feature-level testing was carried out. The system uses three features: Answer–Knowledge cosine similarity (AK) to measure factual grounding, Answer–Question cosine similarity (AQ) to measure relevance, and Length Ratio (LR) to detect abnormal verbosity or under-informative answers. These features were tested both individually and in combination using ablation studies.

A structured ablation study was conducted by training the model with different feature combinations. The full feature set (AK + AQ + LR) achieved the best performance, with an F1-score of 0.75 and balanced precision and recall for both hallucinated and factual answers. When the Length Ratio feature was removed (AK + AQ), the F1-score dropped to 0.69, showing that answer length provides useful supporting information. Using AK + LR further reduced performance to 0.67, indicating that question relevance is also important. Using AK alone resulted in an accuracy of 0.50, which is close to random guessing.

These results confirm that hallucination detection cannot rely on factual similarity alone. While Answer–Knowledge similarity is the strongest signal, hallucinations are best detected when faithfulness, relevance, and answer behaviour are considered together. The observed behaviour matches theoretical expectations and confirms the correctness of the implemented features.

Third, a validation dataset was used to evaluate model performance before testing. The validation results showed slightly lower performance than the test set, which is expected and indicates correct generalisation rather than overfitting.

Finally, the test set was used only once after model selection. Performance was evaluated using precision, recall, F1-score, and confusion matrix, which are appropriate metrics for a binary hallucination detection task.

**2. Performance Evaluation and Interpretation**

The implementation uses:

Sentence-BERT (MiniLM-L3-v2) to generate semantic embeddings,

Gradient Boosting Classifier for hallucination classification.

Sentence-BERT was chosen because hallucinations are semantic errors rather than surface-level text errors. Traditional keyword-based or rule-based methods would fail to capture meaning similarity. MiniLM-L3-v2 was selected because it provides strong semantic representations while remaining computationally efficient, making it suitable for academic experimentation.

Gradient Boosting was chosen instead of deep neural networks because:

The feature space is small and structured,

It handles non-linear feature interactions well,

It performs strongly on tabular data with limited overfitting,

It is easier to interpret and justify than deep end-to-end models.

More complex techniques such as fine-tuning large language models were not used because they require high computational resources and reduce interpretability, which was not suitable for this task.

Results and effectiveness

The final model achieved:

Strong F1-scores on the held-out test set,

Balanced precision and recall,

Low false negatives, which is especially important in medical contexts.

The confusion matrix showed that most hallucinated answers were correctly detected, while factual answers were rarely misclassified. This demonstrates that the system is effective at identifying hallucinations grounded in missing or weak evidence.

Ablation study insights

An ablation study was conducted to evaluate feature importance:

Feature Combination	Result
AK + AQ + LR	Best performance
AK + AQ	Slight performance drop
AK only	Largest performance drop

This shows that:

Answer - Knowledge similarity is the strongest signal, confirming that hallucinations are primarily a lack of factual grounding.

Answer - Question similarity improves relevance checking.

Length ratio provides additional but smaller gains.

These results empirically validate the design choices made in the implementation.

Domain interpretation (Medical QA)

In the medical domain, hallucinations are dangerous because answers may appear fluent but lack evidence. The results show that the system effectively detects such cases by penalising answers that are not semantically aligned with trusted medical knowledge.

This aligns with real clinical risks, where incorrect but confident-sounding answers can mislead users. The system therefore provides a practical safety mechanism for medical AI applications.
In addition to accuracy and F1-score, the model was also evaluated using a Receiver Operating Characteristic (ROC) curve. The ROC curve shows how well the model can tell the difference between hallucinated answers and factual answers at different decision thresholds.

The curve compares how often hallucinated answers are correctly detected with how often factual answers are wrongly marked as hallucinated. This is important in medical systems, where both missing a hallucination and giving a false warning can be risky.

The model achieved an AUC of 0.81, which means it can correctly rank hallucinated answers higher than factual ones in about 81% of cases. Because the curve is well above the random baseline, it shows that the model is learning useful patterns and is not guessing.

**Insights**

Faithfulness is more important than fluency in medical QA.
Semantic grounding against knowledge is a strong hallucination signal.
Lightweight models can achieve strong performance without expensive fine-tuning.

These insights can be used to:
Add hallucination filters to medical chatbots,
Flag unsafe answers for human review,
Improve trust and safety in deployed systems.


**3. Strengths and Limitations**

**Strengths**

One main strength of this implementation is that it uses semantic similarity instead of simple word matching. By using Sentence-BERT embeddings, the model compares the meaning of the answer with both the question and the supporting medical knowledge. This is important because hallucinated answers often sound fluent but are not based on facts.

Another strength is the use of Answer - Knowledge similarity as the most important feature. This directly checks whether the answer is supported by medical evidence. The ablation study showed that removing this feature caused the largest drop in performance, which proves that factual grounding is the key signal for hallucination detection.

The Gradient Boosting classifier works well with a small number of meaningful features and can learn non-linear patterns. Compared to deep neural networks, it is easier to train, more stable, and less likely to overfit.



**Limitations**

A key limitation is that computing sentence embeddings takes time. The model needs to encode questions, answers, and knowledge separately, which increases runtime and makes real-time use more difficult.

The model also does not measure confidence or uncertainty. It only predicts whether an answer is hallucinated or not. In medical systems, it would be safer to know when the model is unsure so that those cases can be reviewed by humans.

Another limitation is that the system uses binary classification. Hallucinations can have different levels of severity, but the current model treats all hallucinations the same.

The model also depends on the quality of the medical knowledge provided. If the knowledge is incomplete or incorrect, the system may wrongly classify a correct answer as a hallucination.



**Future improvements**

Uncertainty estimation can be added to improve safety. The current model makes a hard decision (hallucinated or not), but some predictions are uncertain, especially when similarity scores are moderate. Using prediction probabilities or entropy-based measures would allow the system to flag low-confidence answers and send risky medical responses for human review. This would enhance safety without replacing the existing method.

Third, the binary classification setup could be extended to multi-class classification to represent different levels of hallucination severity, providing more detailed feedback in medical applications.

Fourth, additional signals such as retrieval confidence scores could be combined with the existing features to improve robustness when supporting evidence is weak.

Finally, performance could be improved by testing on more medical QA datasets or by fine-tuning embeddings on medical domain text to improve generalisation.

Conclusion

Overall, the testing and evaluation demonstrate that the implementation is effective, and appropriate for the medical hallucination detection problem. The results align with theoretical expectations, provide meaningful domain insights, and offer a strong foundation for future improvements.

# **PART E - EDI and sustainability**

Equity, Diversity, Inclusion (EDI) and Sustainability
Bias and Fairness

Bias is a real issue in medical AI because training data often reflects gaps in medical research. The MedHallu dataset is based on biomedical literature and expert-written questions, which helps with accuracy but may under-represent minority groups, rare diseases, and non-Western healthcare settings.

This system reduces some bias by not generating medical advice. Instead, it checks whether an answer is supported by medical evidence using Answer–Knowledge semantic similarity. Since it does not use patient demographics, the risk of direct discrimination is low.

However, fairness issues can still happen. Answers about under-researched conditions may be marked as hallucinated simply because less evidence exists. In real clinical use, this would require more diverse medical sources and fairness checks across different medical areas.

Accessibility and Inclusion

The system is a lightweight safety layer that can be added to existing medical QA systems. It uses Sentence-BERT and Gradient Boosting, so it runs on CPUs and does not need GPUs. This makes it suitable for smaller hospitals and low-resource settings.

The output is simple and easy to understand (hallucinated or factual), which helps clinicians trust the system. At the moment, it only works with English medical text. Using multilingual embeddings in the future would improve accessibility worldwide.

Ethical Implications

Medical hallucinations are dangerous because confident but wrong answers can cause harm. This system helps reduce that risk by flagging answers that are weakly supported by medical knowledge, following the principle of do no harm.

It is designed to support clinicians, not replace them. Risky answers are flagged for human review. However, missing a hallucination is still a concern, so ethical use requires cautious thresholds, human oversight, and clear communication that predictions are uncertain.

Environmental Sustainability

Compared to large language models, this system is computationally efficient. Sentence-BERT MiniLM provides good semantic performance at low cost, and Gradient Boosting runs quickly on CPUs.

Embedding generation is the main cost, but it is done once and does not scale with model size. Storage needs are small, as only text and compact embeddings are stored.

Environmental impact can be reduced by caching embeddings, using CPU-only inference, and applying the system only to high-risk medical queries. Preventing unsafe hallucinations also reduces follow-up errors and unnecessary treatments.

Overall Evaluation

Overall, this system follows responsible AI practices in healthcare. It tackles medical uncertainty using simple, interpretable, and efficient methods. While there are limits around data bias and language support, it provides a practical and ethical safety layer for medical decision-support systems.
